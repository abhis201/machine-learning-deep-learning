# Q2

```python
def right_push(arr, push, num):
    arr = np.roll(arr,push)
    for i in range(push):
        arr[i] = num
    return arr
```

```python
a = np.arange(1,11)
a
```

```python
b = a
b = right_push(b,1,0)
b
```

```python
a.size==b.size
```

```python
print("a_shape = "+str(a.shape))
print("b_shape = "+str(b.shape))
```

# Q3

```python
def create_mask_add(arr, mask):
    b = np.tril(arr)
    np.fill_diagonal(b,0)

    return (b - arr)
```

```python
x = np.random.randint(1,10,size = (3,3))
x
```

```python
mask = create_mask_add(x, 0)
mask
```

```python
new_x = x + mask
new_x
```

```python
mask_mul = np.array([[0,0,0],[1,0,0],[1,1,0]])
mask_mul
```

```python
new_x = x * mask_mul
new_x
```
